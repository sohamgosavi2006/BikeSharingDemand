```python
# Procedure -> Converting Numerical values of column table to String

# Assigning 'string'(name of season) values to different seasons instead of numeric values

# 1 -> Spring , 2 -> Summer , 3 -> Fall , 4 -> Winter
# .loc[] is used to select data by labels (row labels and column names).
dataset.loc[(dataset['season'] == 1), 'season'] = 'Spring'
dataset.loc[(dataset['season'] == 2), 'season'] = 'Summer'
dataset.loc[(dataset['season'] == 3), 'season'] = 'Fall'
dataset.loc[(dataset['season'] == 4), 'season'] = 'Winter'

# 0 -> 2018 , 1 -> 2019 (735 rows = 365 + 365 days)
dataset.loc[(dataset['yr'] == 0), 'yr'] = 2018
dataset.loc[(dataset['yr'] == 1), 'yr'] = 2019
```

```python

dataset['season'].astype('category').value_counts()
```

```python
# Assigning string values to month [name of month] instead of numerical values

# dataset.loc[(dataset['mnth'] == 1), 'mnth'] == 'JAN'

# Creating using function
# 'x' is panda series from 1 to 12
def object_map_months(x):
  return x.map({1 : "JAN" , 2: "FEB",3 : "MAR" , 4: "APR",5 : "MAY" , 6: "JUN",
                7 : "JUL" , 8: "AUG",9 : "SEP" , 10: "OCT",11 : "NOV" , 12: "DEC"})

# Applying Function to Data
dataset[['mnth']] = dataset[['mnth']].apply(object_map_months);
```

```python
# Checking updated data for months
dataset['mnth'].astype('category').value_counts()
```

```python
# Assigning values [numerial to string] for Weekdays
def object_map_weekdays(x):
  return x.map({0 : "SUN", 1 : "MON" , 2: "TUE", 3 : "WED" , 4: "THU",5 : "FRI" , 6: "SAT"})

dataset[['weekday']] = dataset[['weekday']].apply(object_map_weekdays)
```

```python
# Checking if values is equally distributed for days of week for entire dataset
dataset['weekday'].astype('category').value_counts()
```

```python
# For Weather -
# 1-> Clear , 2-> Mist , 3-> Rain

dataset.loc[(dataset['weathersit'] == 1), 'weathersit'] = 'Clear'
dataset.loc[(dataset['weathersit'] == 2), 'weathersit'] = 'Mist'
dataset.loc[(dataset['weathersit'] == 3), 'weathersit'] = 'Rain'
```

```python
# We would REMOVE atemp(actual temperature) to avoid repititionas as it is highly
# corelated with temp which means it is nearly giving the same result as temp
# eg -> temp =30 and atemp =29.8

dataset.drop('atemp',axis=1,inplace=True)
dataset.head()
```

```python
# @title 3.) DATA PREPARATION
```

```python
dataset_categorical = dataset.select_dtypes(include = ['object'])
dataset_categorical.head()
```

```python
# Creating DUMMY DATA [Dummy Variables]

# Concept ->
# You take one original column (eg -> Season)
# Each new dummy column[spring,summer,winter] asks a yes/no question w.r.t Seasons column
# "Is this row's Season Winter?" â†’ If yes, put 1, if no, put 0.
# "Is this row's Season Summer?" â†’ If yes, put 1, if no, put 0.

# It helps to convert categorical columns into a format that a machine learning model can understand (1-> True , 0 -> False)
dataset_dummies = pd.get_dummies(dataset_categorical,drop_first=True)
dataset_dummies.head()
```

```python
# Drop Categorical variable columns : removing season, month, weekday, weather column
dataset = dataset.drop(list(dataset_categorical.columns),axis=1)
dataset.head()
```

```python
# Now Concatenate Dummy Variables with the Dataset

# axis = 1 -> In pandas means operate along Columns
dataset = pd.concat([dataset,dataset_dummies],axis=1)

# To train the model better and for more accuracy, it is necessary to remove unwanted data
# Dropping extra information (yr column is already present so no need to include date-month-day column)
dataset = dataset.drop(['instant', 'dteday'] , axis =1)
dataset.head()
```

```python
# @title 4.) MODEL BUILDING AND EVALUATION
```

```python
from sklearn import linear_model
from sklearn.linear_model import LinearRegression

# Rows are shuffled and splitted into ->
# df_train â†’ the training data (70% of the rows)
# df_test â†’ the testing data (30% of the rows)

# Training data â†’ used by the model to learn patterns.
# Testing data â†’ used after training to check how well the model performs on data it hasnâ€™t seen before.

# Split the Dataframe into Train and Test Dataset
from sklearn.model_selection import train_test_split
np.random.seed(0)

# Spliting data into Training data and Testing data
# random_state will fix the shuffled rows and every time we run code, same split is followed and
# on removing random_state, the rows picked for training/testing may change every run, making results harder to compare.
df_train, df_test = train_test_split(dataset, train_size = 0.7, test_size = 0.3, random_state = 100)

df_train
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

# Many machine learning models (like KNN, neural networks) work better when all features are on the same scale.
# scaledÂ value = (value - min)/(max - min) [after scaling all values are between 0 to 1 scale range]

# Apply scaler to all columns except Dummy Variables [ only 0/1 value (yes/no) ]
var = ["temp" , "hum" , "windspeed" , "casual" , "registered" , "cnt"]
df_train[var] = scaler.fit_transform(df_train[var])
df_train.head()
```

```python
# To check corelation coefficient to see which variables (two) are highly similar to each other
plt.figure(figsize=(25,25))
sns.heatmap(df_train.corr(),annot=True,cmap ="YlGnBu")
plt.show()
```

```python
# X â†’ Our features (input variables)
# This is the data the model uses to make predictions
# Example: temperature, humidity, windspeed

# y â†’ Your target (output variable) [it is what we are trying to predict]
# This is what we want the model to predict
# Example: bike rental count (cnt)

# Regenerating df_train from the original dataset before popping 'cnt'
df_train, df_test = train_test_split(dataset, train_size = 0.7, test_size = 0.3, random_state = 100)
var = ["temp" , "hum" , "windspeed" , "casual" , "registered" , "cnt"]
scaler = MinMaxScaler()
df_train[var] = scaler.fit_transform(df_train[var])

y_train = df_train.pop('cnt')
X_train = df_train.drop(['casual' , 'registered'], axis = 1)


# Convert boolean (True/False) columns to integers
for col in X_train.columns:
    if X_train[col].dtype == 'bool':
        X_train[col] = X_train[col].astype(int)

X_train.head()
```

```python

np.array(X_train)
```

```python
import statsmodels.api as sm

# Adds a column of 1â€™s to our training data so the regression can calculate the
# intercept (the point where the line crosses the Y-axis) or else it would pass through origin
X_train_lm = sm.add_constant(X_train)

# OLS -> Ordinary Least Squares [Method to find best fitting line through Data]
lr = sm.OLS(y_train, X_train_lm).fit()
lr.params

# Example ->

# const         50.2
# temp          20.3
# hum           -5.1

# Intercept = 50.2
# For every 1Â° increase in temp, prediction increases by +20.3 (keeping other data(hum) constant)
```

```python
# Creating Linear Regression

# Linear regression in ML is one of the simplest algorithms used to predict a value based on other values.

# We draw a straight line through our data points so that the line is as close as possible to all points and
# That line can then be used to make predictions

lm = LinearRegression()
lm.fit(X_train,y_train)
```

```python
print(lm.coef_)
print(lm.intercept_)
```

```python
# If R - value comes 1 (error in model)
# R-squared = 0.853 â†’ our model explains 85% of the changes in bike rentals and other 15% is due
# to other factors which our model cannot predict
lr.summary()

# R-squared (RÂ²):
# It tells you how much of the changes in your target (y) can be explained by your modelâ€™s inputs (X).

# RÂ² = 1 (or 100%) â†’ Your guesses match perfectly.
```

```python
# Using Recursive Feature Elimination (RFE)
from sklearn.feature_selection import RFE

# RFE will help train the model by removing the least important features

lm = LinearRegression()
# Tells RFE to find 15 best features from
# Estimator is the model RFE will use to decide which features are important (here LinearRegression Model is used)
rfe1 = RFE(estimator=lm, n_features_to_select=15)

# Fitting with 15 best features
rfe1.fit(X_train,y_train)

print(rfe1.ranking_)
print(rfe1.support_)
```

```python
# Printing best 15 Features according to Linear Regression Model from dataset column
col1 = X_train.columns[rfe1.support_]
col1
```

```python
X_train_rfe1 = X_train[col1]

X_train_lm = sm.add_constant(X_train_rfe1)
lm1 = sm.OLS(y_train, X_train_lm).fit()
lm1.summary()

# Adj. R-squared value has dropped by 0.1 percent only from the previous measurement where we had more 14 columns
# This shows that out model is performing good by optimising resources provided to producing accurate result
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

# VIF -> VIF = 1 : No Overlap , 5 > VIF > 1 : Some Overlap , VIF > 5 : High Overlap
# It tells us how much one feature (column) is overlapping / correlated
# with other features in the dataset.

# Purpose of VIF -> If two features (temp and atemp) are almost the same, the model gets confused hence
#                   VIF helps detect and remove duplicate features for model optimisation


vif = pd.DataFrame()
vif['features'] = X_train_rfe1.columns
vif['VIF'] = [variance_inflation_factor(X_train_rfe1.values, i) for i in range(X_train_rfe1.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

```python
# lm -> Linear Model
lm = LinearRegression()

rfe2 = RFE(estimator=lm, n_features_to_select=7)

# Fitting with 7 best features
rfe2.fit(X_train,y_train)

print(rfe2.ranking_)
print(rfe2.support_)
```

```python
# Printing Top 7 features of column according to Linear RegressioN Model
col2 = X_train.columns[rfe2.support_]
col2
```

```python
X_train_rfe2 = X_train[col2]

X_train_lm = sm.add_constant(X_train_rfe2)
lm2 = sm.OLS(y_train, X_train_lm).fit()
lm2.summary()

# Adj. R Squared -> 81.1%
```

```python
vif1 = pd.DataFrame()
vif1['features'] = X_train_rfe2.columns
vif1['VIF'] = [variance_inflation_factor(X_train_rfe2.values, i) for i in range(X_train_rfe2.shape[1])]
vif1['VIF'] = round(vif1['VIF'], 2)
vif1 = vif1.sort_values(by = "VIF", ascending = False)
vif1

# Again hum and temp has highest Variance [shows how one feature is simila to all other features in dataset]
```

```python
# Add a constant to X_train_rfe2 before prediction
X_train_rfe2_lm = sm.add_constant(X_train_rfe2)
y_train_cnt = lm2.predict(X_train_rfe2_lm)

# To find how much our model is wrong for each prediction ->
# y_train = actual values (real bike rentals)
# y_train_cnt = predicted values (modelâ€™s guesses)

# Subtracting them â†’ Errors  = actual âˆ’ predicted

# Bins -> When plotting a histogram, we divide the data into intervals which are bins
fig = plt.figure()
sns.distplot(y_train - y_train_cnt, bins = 20)
fig.suptitle('Error Terms', fontsize = 20)
plt.xlabel('Errors', fontsize = 18)
```

```python
# Regenerate df_test from the original dataset
df_train, df_test = train_test_split(dataset, train_size = 0.7, test_size = 0.3, random_state = 100)

# Scaling the test data
var = ["temp" , "hum" , "windspeed" , "casual" , "registered" , "cnt"]
df_test[var] = scaler.transform(df_test[var])

# Separating features and target variable from the test set
y_test = df_test.pop('cnt')
X_test = df_test.drop(['casual' , 'registered'], axis = 1)

# Convert boolean columns to integers in X_test
for col in X_test.columns:
    if X_test[col].dtype == 'bool':
        X_test[col] = X_test[col].astype(int)

# Select the same RFE-selected features from X_test
X_test_rfe2 = X_test[col2]

# Add a constant to X_test_rfe2
X_test_rfe2_lm = sm.add_constant(X_test_rfe2)

# Make predictions on the test set
y_pred = lm2.predict(X_test_rfe2_lm)

# Evaluate the model
from sklearn.metrics import r2_score
r2 = r2_score(y_test, y_pred)
print(f"R-squared on the test set: {r2}")

# 80.5 % is R- squared
```
